## Accuracy vs SNR

```python
labels = ['CNN', 'LSTM', 'CNN-LSTM', 'CNN-S', 'LSTM-S', 'CNN-LSTM-S']
colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:cyan', 'tab:olive']
markers = ['+' , 'x', 'd', '*' , 'h', 'o']
subfixes = ['', '-FDA', '-FDA+']
lines = ['-', '--', '-.']
x_labels = ['10 dB', '15 dB', '20 dB', 'No Noise']

for i in range(len(labels)):
    l = labels[i]
    c = colors[i]
    m = markers[i]

    for j in range(len(subfixes)):
        s = subfixes[j]
        ls = lines[j]
        style = ls + m
        key = l + s
        x_value = []
        x_value.append(float("{:.3f}".format(results_snr10[key])))
        x_value.append(float("{:.3f}".format(results_snr15[key])))
        x_value.append(float("{:.3f}".format(results_snr20[key])))
        x_value.append(float("{:.3f}".format(results[key])))

        plt.plot(x_value, style, color=c, label=key)

plt.title('Signal classification accuracy vs. SNR over an AWGN channel models')
plt.xlabel('Noise')
plt.ylabel('Classification Accuracy')
plt.xticks(np.arange(len(x_labels)), x_labels)  # Set text labels.
plt.legend(bbox_to_anchor=(1, 1), loc='upper left')
plt.show()
```
