## No Noise

```python
results = {'FCNN': 0.5738974301425205, 'FCNN-FDA': 0.5838850858489507, 'FCNN-FDA+': 0.5910672202895298, 
           'CNN': 0.9374929861968354, 'CNN-FDA': 0.9608349231287173, 'CNN-FDA+': 0.9051733812142296, 
           'CNN-S': 0.6876122082585279, 'CNN-S-FDA': 0.9730700179533214, 'CNN-S-FDA+': 0.9290843806104129, 
           'LSTM': 0.8200179533213644, 'LSTM-FDA': 0.6490125673249552, 'LSTM-FDA+': 0.8258527827648114, 
           'LSTM-S': 0.6879138143867131, 'LSTM-S-FDA': 0.6707440242397037, 'LSTM-S-FDA+': 0.6883626977892492, 
           'CNN-LSTM': 0.9433284704298058, 'CNN-LSTM-FDA': 0.9685781618224666, 'CNN-LSTM-FDA+': 0.931881943665133, 
           'CNN-LSTM-S': 0.9600538599640933, 'CNN-LSTM-S-FDA': 0.9649910233393177, 'CNN-LSTM-S-FDA+': 0.9456912028725314}

other_results = {'CNN-1D': 0.8679160588037257, 'CNN-W': 0.9466950959488273}
```

```python
labels = ['FCNN', 'CNN', 'LSTM', 'CNN-LSTM', 'CNN-S', 'LSTM-S', 'CNN-LSTM-S']

without_fda = []
fda = []
fda_plus = []

for l in labels:
    without_fda.append(float("{:.3f}".format(results[l])))
    fda.append(float("{:.3f}".format(results[l + '-FDA'])))
    fda_plus.append(float("{:.3f}".format(results[l + '-FDA+'])))

x = np.arange(len(labels))
width = 0.3

fig, ax = plt.subplots()
rects_seg = ax.bar(x - width, without_fda, width, label='Segments', edgecolor='black')
rects_1d = ax.bar(6.75, float("{:.3f}".format(other_results['CNN-1D'])), width, edgecolor='black')
rects_w = ax.bar(7.3, float("{:.3f}".format(other_results['CNN-W'])), width, edgecolor='black')
rects_fda= ax.bar(x, fda, width, label='FDA', edgecolor='black')
rects_fdap = ax.bar(x + width, fda_plus, width, label='FDA+', edgecolor='black')


ax.set_ylabel('Classification Accuracy')
ax.set_title('Classification Accuracies without Noise by Architecture and Input Data', pad=15)
ax.set_xticks(np.concatenate((np.arange(len(labels)), np.array([6.75, 7.3]))))
ax.set_xticklabels(labels + ['CNN-1D', 'CNN-W'])
ax.set_xlabel('Architectures')
ax.legend(title='Input Data')

ax.bar_label(rects_seg, label_type='edge', rotation=45)
ax.bar_label(rects_fda, label_type='edge', rotation=45)
ax.bar_label(rects_fdap, label_type='edge', rotation=45)
ax.bar_label(rects_1d, label_type='edge', rotation=45)
ax.bar_label(rects_w, label_type='edge', rotation=45)

fig.tight_layout()
plt.show()

max_key = max(results.items(), key=operator.itemgetter(1))[0]
print(max_key, results[max_key])
```
